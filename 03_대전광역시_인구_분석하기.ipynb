데이터 준비하기

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df=pd.read_csv('/content/200812_202112_주민등록인구및세대현황_대전.csv',encoding='cp949')
df.head()
```

```python
df.info()
```

데이터 전처리

```python
daejeon_df = df.loc[:,['행정구역',
                       '2008년_총인구수',
                       '2009년_총인구수',
                       '2010년_총인구수',
                       '2011년_총인구수',
                       '2012년_총인구수',
                       '2013년_총인구수',
                       '2014년_총인구수',
                       '2015년_총인구수',
                       '2016년_총인구수',
                       '2017년_총인구수',
                       '2018년_총인구수',
                       '2019년_총인구수',
                       '2020년_총인구수',
                       '2021년_총인구수']]
daejeon_df
```

```python
daejeon_df.info()
```

```python
dj_df = daejeon_df.set_index('행정구역').T
dj_df
```

```python
dj_df.reset_index(drop=False)
```

```python
dj_df.to_csv('/content/행정구역별_년도별총인구수.csv',encoding='cp949',index_label='연도')
```

시각화로 보는 인구 이동 및 감소/증가

```python
fig = px.bar(daejeon_df,
             x="행정구역",
             y=daejeon_df.columns,
             barmode='group',
             text_auto='True',
             title="2008년~2021년까지 대전 인구수 데이터 분석")

fig.show()
```

```python
daejeon_total_df = pd.read_csv("/content/행정구역별_년도별총인구수.csv",encoding='CP949')
```

```python
fig=px.line(daejeon_total_df,
             x="연도",
             y=['동구','중구','서구','유성구','대덕구'],)

fig.update_layout(title='2008년~2021년 대전 총인구수 데이터 분석',
                  xaxis_title='2008년~2021년',
                  yaxis_title='대전광역시 구별 인구수',
                  autosize=False,
                  width=500,height=500)
```
